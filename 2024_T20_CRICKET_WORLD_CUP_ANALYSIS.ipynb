3.Most Economical Bowlers

```python
total_runs = df1.groupby('bowler')['runs_off_bat'].sum().reset_index()
total_runs

total_balls=df1.groupby('bowler')['ball'].count().reset_index()
total_balls

Economy_rate_df=pd.merge(total_runs,total_balls,on='bowler')
Economy_rate_df=Economy_rate_df.rename(columns={'runs_off_bat':'runs_conceded','ball':'Balls Bowled'})
Economy_rate_df

```

```python
economy_rate = Economy_rate_df.groupby('bowler')['runs_conceded'].sum() / (Economy_rate_df.groupby('bowler')['Balls Bowled'].sum() / 6)
economy_rate =economy_rate.sort_values(ascending=True).head(10).reset_index().rename(columns={0:'Economy Rate'})
economy_rate['Economy Rate'] = economy_rate['Economy Rate'].round(2)
economy_rate
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=economy_rate.head(10),y='bowler',x='Economy Rate',hue='bowler',palette='dark:red' , orient='h' )
plt.ylabel('Bowler')
plt.xlabel('Economy Rate')
plt.title('Top 10 Economical Bowler in 2024 T20 WC ( LOWER THE BETTER )')
plt.show()
```

**OBSERVATIONS**
* **TG Southee** , **R Ravindra** and **Mujeeb Ur Rahman** are the top 3 most economical bowlers in this 2024 T20 WC.
